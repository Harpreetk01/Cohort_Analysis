## Cohort Analysis

```python
d1 = data[['CustomerID', 'InvoiceDate']]
d1
```

```python
# Extract month of the first purchase
d1['Cohort_Month'] = d1.groupby('CustomerID')['InvoiceDate'].transform('min')
d1['Cohort_Month'] = d1['Cohort_Month'].dt.to_period('M')
data1 = d1.drop(['Cohort_Month'], axis = 1)
```

```python
data1
```

```python
# Assign each customer to a cohort
data1['TransactionMonth_index'] = data1['InvoiceDate'].dt.to_period('M')
data1['Cohort_Index'] = data1['TransactionMonth_index'] - data1['CohortMonth']
data1['Cohort_Index'] = data1['Cohort_Index'].apply(lambda x: x.n)
data1
```

```python
# Calculate monthly active customers per cohort
cohort = data1.groupby(['CohortMonth', 'Cohort_Index'])['CustomerID'].nunique().reset_index()
cohort.rename(columns={'CustomerID': 'TotalUsers'}, inplace=True)
```

```python
cohort
```

```python
# Create pivot table with retention rate
cohort_pivot = cohort.pivot(index = 'CohortMonth', columns = 'Cohort_Index', values = 'TotalUsers')
cohort_pivot
```

```python
cohort_size = cohort_pivot.iloc[:,0]
retention_matrix = cohort_pivot.divide(cohort_size, axis=0).round(3)
retention_matrix
```

```python
# Visualize retention matrix
plt.figure(figsize=(12,10))
plt.title('Retention Rate by Monthly Cohorts')
sns.heatmap(data=retention_matrix, annot=True, fmt='.0%', cmap='YlGnBu', vmin=0, vmax=0.5)
plt.show()
```
